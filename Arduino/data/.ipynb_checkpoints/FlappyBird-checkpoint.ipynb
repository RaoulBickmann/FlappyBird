### Aufnahme mit 10 cm von wand entfernt und anschlieÃŸendes plotten

```python
def readData():
    data = pd.read_csv('10sec10cm')
    data.time /= 1000
    data.time -= data.time[0]
    data = data[data.time>=0]
    data = data[data.time<= 11000]
    
    std = data.distance.std()
    var = std ** 2
    mean = data.distance.mean()
    
    print(mean)
    print(std)
    
    data = data[data.distance>=mean - std]
    data = data[data.distance<= mean + std]
    
    return data
```

```python
def plot(columns):
    plt.figure(figsize=(20, 20))
    for idx,name in enumerate(columns):
        plt.subplot(3,1,idx+1)
        plt.plot(data.time, data[name], color='rgb'[idx], label=name)
        plt.legend(loc='upper left')
        plt.grid()
        plt.xticks(range(0,12, 1))

data = readData()
plot(['distance'])
```
